## 1.1 Churn in the Telcommunications industry

In telecommunications industry, customers are able to choose among multiple service providers and actively exercise their right of switching from one service provider to another. 

In this strongly competitive market,customers demand better services at less price, while service providers constantly focus on retention as their business goal. In this context it is important to manage the phenomenon of churn, i.e. the propensity of customers to cease doing business with a company in a given time period. 

Given the fact that the telecommunications industry experiences high churn rate and it costs more to recruit a new customer than to retain an existing one, customer retention is clearly more important than customer acquisition.

Many telecommunications companies apply retention strategies to keep customers longer. With retention strategies in place, many companies start to include churn reduction as one of their business goals. A good prediction of the next leaving customer, leads marketing managers to conceive targeted retention strategies.

## 1.2 The Purpose of the Experiment : Predicting Churn in Telecom Industry

Multi-layer perceptrons are complex artificial neural networks (ANN). Among the most common problems addressed with the use of ANNs is worth mentioning:

1. Binary Classification
2. Multiclass Classification
3. Regression

Binary classification is a type of supervised learning, a method of machine learning where the categories are predefined, and is used to categorize new probabilistic observations into said categories. When there are only two categories the problem is known as binary classification.

We wish to create an ANN able to correctly classify as many instances of customer churn in a telecom operator as possible. 

For this reason, an telecom data set freely available on the internet and including the necessary variables to perform churn propensity calculation has been downloaded - further detail about the data set can be found in the rest of the document.

As a minimum, we expect +50% of the churn instances to be correctly classified. A model that returns an accuracy less or equal than 50% has certainly poor predictive power - to the point that tossing a coin might give higher probability to correctly predict the event than using the algorithm generated.

A model that hits the minimum target of +50% accuracy and achieves a moderately high accuracy might be more or less useful. This because investing resources to retain customers who are, in reality, not likely to churn represents a cost opportunity and might lead to revenue loss for the telecom operator.

On the contrary, a high accuracy in distinguishing probable churners from non-churners would allow to put in place marketing actions to retain and re-engage customers who are intended to churn.

## 2.3.2 Handling Categorical Features :  One hot encoding

Categorical data are variables that contain label values rather than numeric values. 
The number of possible values is often limited to a fixed set.

What is the Problem with Categorical Data?

Some algorithms can work with categorical data directly. However, not any machine learning algorithms can operate on label data directly.
They require all input variables and output variables to be numeric. This is where one hot encoding plays its important role.

A one hot encoding is a representation of categorical variables as binary vectors.
This first requires that the categorical values be mapped to integer values.
Then, each integer value is represented as a binary vector that is all zero values except the index of the integer, which is marked with a 1.

In general, this is mostly a constraint of the efficient implementation of machine learning algorithms rather than hard limitations on the algorithms themselves. 
This means that categorical data must be converted to a numerical form. 

If the 	categorical variable is an output variable, we could want to convert predictions by the model back into a categorical form in order to present them or use them in some application.

Note that for categorical variables where no ordinal relationship exists, the integer encoding used for ordinal avriables is not enough. 
In fact, using this encoding and allowing the model to assume a natural ordering between categories may result in poor performance or unexpected results.

Behind the complexity of an ANN there is a linear function smoothed throguh a sigmoidal activation function. The target, or label, in an ANN is what we aim to classify or predict.

It is good practice to separate out features and targets, as this latter will be the dependent variable we are trying to handle and the former will be the data from which we will try to extract the function that best explains the value of the target variable.

### It is good practice to shuffle the data prior to build the ANN

Unless we are dealing with time-ordered data, it is good practice to shuffle the data before splitting it into training, validation and test sets. 

This is a way to avoid obtaining entire minibatches of highly correlated examples.

Shuffling the data prior to any split helps to reduce the risk of unrepresentativeness of the validation and test data sets.
In an extreme case where delaing with data sorted in a certain order, splitting the data according to the row number in the data set would mean for the ANN to train to classify the targets based on data points that are rare or do not exist at all in the other sets - this effect woudl also depend on the variance in the data

### 2.3.4 Splitting the Data into Training and Test Data Sets

One of the very common issues while developing Machine Learning systems is overfitting.

When you train your model, it tries its best to find out some kind of pattern in your training data while minimizing the error rate. But we only want our model to find pattern. Not more than that.
This because if your model just tries to go beyond finding patterns, it may memorize the training data. This is one of the reasons we need to use training and test set separation. 

After we have trained our model, we should test it on fresh examples so that we can avoid problems like overfitting. 
We can also estimate how well our model is performing given that it is facing new inputs. Based on the performance, we can proceed on developing our system further.

So, the test data set is nothing but the sample of data used to provide an unbiased evaluation of a final model fit on the training dataset.

The point is, good accuracy in training set can ensure nothing.


```python
def trn_tst_spl(trn):
    global x_train
    x_train=DATA[:(int(round((np.shape(DATA)[0]*(trn)),0)))]
    global y_train
    y_train=TARGETS[:(int(round((np.shape(DATA)[0]*(trn)),0)))]
    global x_test
    x_test=DATA[:(int(round((np.shape(DATA)[0]*(1-trn)),0)))]
    global y_test
    y_test=TARGETS[:(int(round((np.shape(DATA)[0]*(1-trn)),0)))]
```

### 2.3.5 Splitting Training Data Set into Partial Training and Validation Data Sets

As a good accuracy in training set can ensure nothing about the model application on the real data, a more encouraged technique is to divide your data set into three classes: training set, cross validation set and test set. 

The validation set is used to evaluate a given model, but this is for frequent evaluation. Specifically, we can use this data to fine-tune the model hyperparameters. Hence the model occasionally sees this data, but never does it learn from this. We use the validation set results and update higher level hyperparameters. So the validation set in a way affects a model, but indirectly.

How to split the dataset into Train, Validation and Test sets mainly depends on 2 things. 
First, the total number of samples in your data and second, on the actual model you are training.

Some models need substantial data to train upon, so in this case you would optimize for the larger training sets. 
Models with very few hyperparameters will be easy to validate and tune, so you can probably reduce the size of your validation set, but if your model has many hyperparameters, you would want to have a large validation set as well (although cross validation might be considered as well). 

All in all,  the train-test-validation split ratio is also quite specific to your use case and it gets easier to make judgement as more and more models are trained and built.


```python
def trn_vld_spl(trn):
    global partial_x_train
    partial_x_train=x_train[:(int(round((np.shape(x_train)[0]*(trn)),0)))]
    global val_x_train
    val_x_train=x_train[:(int(round((np.shape(x_train)[0]*(1-trn)),0)))]
    global partial_y_train
    partial_y_train=y_train[:(int(round((np.shape(x_train)[0]*(trn)),0)))]
    global val_y_train
    val_y_train=y_train[:(int(round((np.shape(x_train)[0]*(1-trn)),0)))]
```

### 2.3.6 Normalising Training Data Prior to Split into Partial Training and Validation Data Sets

When numeric  x-data values are normalized, neural network training is often more efficient, which leads to a better predictor. 

This because if numeric data is not normalized, and the magnitudes of two predictors are far apart, then a change in the value of a neural network weight has far more relative influence on the x-value with larger magnitudes.

In other words, if the scales for different features are wildly different, this can have a knock-on effect on the ability of the ANN to learn. Ensuring standardised feature values tackles this event to occur.

```python
# TRAINING DATA NORMALISATION

def trn_val_spl_nrm(trn):

    # re-splitting the data into training and test to normalise the training data
    trn_tst_spl((3/4))

    # training data normalisation:
    mean=x_train.mean(axis=0)
    std_dev=x_train.std(axis=0)
    global x_train_nrm
    x_train_nrm=x_train-mean
    x_train_nrm/=std_dev
    global x_test_nrm
    x_test_nrm=x_test-mean
    x_test_nrm/=std_dev
    
    # generating normalised partial training and validation data sets
    global partial_x_train_nrm
    partial_x_train_nrm=x_train_nrm[:(int(round((np.shape(x_train_nrm)[0]*(trn)),0)))]
    global val_x_train_nrm
    val_x_train_nrm=x_train_nrm[:(int(round((np.shape(x_train_nrm)[0]*(1-trn)),0)))]
    global partial_y_train_nrm
    partial_y_train_nrm=y_train[:(int(round((np.shape(x_train_nrm)[0]*(trn)),0)))]
    global val_y_train_nrm
    val_y_train_nrm=y_train[:(int(round((np.shape(x_train_nrm)[0]*(1-trn)),0)))]
```

```python
def model(data_validation,num_epc,cross_validation,num_neu_inp,wre_l1,wre_l2,hlr_act,drp_out,num_neu_hid,
          extra_layer1,lr,los_fun,kpi,xtr_dta,ytr_dta,xvl_dta,yvl_dta,verbose):
    
    global k
    if data_validation=='hold_out':
        k=1
        from tensorflow.keras import models
        global num_epochs
        num_epochs=num_epc
        my_model=models.Sequential()
        from tensorflow.keras import layers
        from tensorflow.keras import regularizers
        my_model.add(layers.Dense(num_neu_inp,kernel_regularizer=regularizers.l1_l2(l1=wre_l1,l2=wre_l2),activation=hlr_act,input_shape=(x_train.shape[1],)))
        my_model.add(layers.Dropout(drp_out))
        my_model.add(layers.Dense(num_neu_hid,kernel_regularizer=regularizers.l1_l2(l1=wre_l1,l2=wre_l2),activation=hlr_act))
        my_model.add(layers.Dropout(drp_out))
        #### Adding an extra hidden layer and the associated drop out
        extra_layer1=extra_layer1
        if extra_layer1==1:
            my_model.add(layers.Dense(num_neu_hid,kernel_regularizer=regularizers.l1_l2(l1=wre_l1,l2=wre_l2),activation=hlr_act))
            my_model.add(layers.Dropout(drp_out))
        
        my_model.add(layers.Dense(1,activation='sigmoid'))
        from tensorflow.keras import optimizers
        from tensorflow.keras import losses
        from tensorflow.keras import metrics
        my_model.compile(optimizer=optimizers.RMSprop(lr=lr),loss=los_fun,metrics=[kpi])

        #print('Below the summary of the parameters processed :'+'\n'+str(my_model.summary()))
        
        from datetime import datetime
        start_time = datetime.now()

        globals()['out']= my_model.fit(xtr_dta,ytr_dta,epochs=num_epochs,batch_size=16, validation_data=(xvl_dta,yvl_dta),verbose=verbose)
        out=(globals()['out'])

        end_time = datetime.now()

        global out_his
        out_his=out.history

        #### Model fit time
        global dur        
        dur=str(end_time - start_time)
 
        #### argmin and argmax
        global max_vld_acc
        global max_trn_acc
        global min_vld_lss
        global min_trn_lss

        max_vld_acc=(np.argmax(out_his['val_acc']))+1
        max_trn_acc=(np.argmax(out_his['acc']))+1
        min_vld_lss=(np.argmin(out_his['val_loss']))+1
        min_trn_lss=(np.argmin(out_his['loss']))+1

        #### mean
        global av_vld_acc
        global av_trn_acc
        global av_vld_lss
        global av_trn_lss

        av_vld_acc=round(np.mean(out_his['val_acc']),2)
        av_trn_acc=round(np.mean(out_his['acc']),2)
        av_vld_lss=round(np.mean(out_his['val_loss']),2)
        av_trn_lss=round(np.mean(out_his['loss']),2)

        #### standard deviation
        global std_vld_acc
        global std_trn_acc
        global std_vld_lss
        global std_trn_lss

        std_vld_acc=round(np.std(out_his['val_acc']),4)
        std_trn_acc=round(np.std(out_his['acc']),4)
        std_vld_lss=round(np.std(out_his['val_loss']),4)
        std_trn_lss=round(np.std(out_his['loss']),4)

        #### Coefficient of correlation (linear, Pearson)
        epochs=list(range(1,len(out_his['loss'])+1))
        val_acc=list(out_his['val_acc'])
        val_loss=list(out_his['val_acc'])
        acc=list(out_his['acc'])
        loss=list(out_his['loss'])

        global kco_val_acc
        kco_val_acc=round(np.corrcoef(epochs,val_acc)[0,1],2)
        global kco_trn_acc
        kco_trn_acc=round(np.corrcoef(epochs,acc)[0,1],2)
        global kco_val_lss
        kco_val_lss=round(np.corrcoef(epochs,val_loss)[0,1],2)
        global kco_trn_lss
        kco_trn_lss=round(np.corrcoef(epochs,loss)[0,1],2)

    elif data_validation=='k_fold': ############################### K-FOLD CROSS VALIDATION:
        k=4
        z=len(x_train_nrm)//k
        
        num_epochs=num_epc
        for i in range(k):
            print('processing fold #',str(i+1))
                    
            cv=cross_validation

            if cv=='itarated': 
                #### perform iterated k-fold cross validation with shuffle
                
                import random
                sample=list(random.sample(range(len(x_train)), z))
                total_range=list(range(len(x_train)))
                not_in_sample = list(set(total_range)-set(sample))
                # validation and training data for interated k-fold validation with shuffle:
                val_data=x_train[sample]
                partial_train_data=x_train[not_in_sample]
                # validation and training labels for interated k-fold validation with shuffle:
                val_targets=y_train[sample]
                partial_train_targets=y_train[not_in_sample]
            
            else: 
                #### perform standard k-fold cross validation
            
                # validation and training data for k-fold validation:              
                val_data=x_train[i*z:(i+1)*z]
                partial_train_data=np.concatenate([
                x_train[:i*z],x_train[(i+1)*z:]],axis=0)
                # validation and training labels for k-fold validation:
                val_targets=y_train[i*z:(i+1)*z]
                partial_train_targets=np.concatenate([
                y_train[:i*z],y_train[(i+1)*z:]],axis=0)

            #### Model build:
            from tensorflow.keras import models
            from tensorflow.keras import layers
            from tensorflow.keras import optimizers
            from tensorflow.keras import regularizers
            from tensorflow.keras import losses
            from tensorflow.keras import metrics
            my_model=models.Sequential()
            my_model.add(layers.Dense(num_neu_inp,kernel_regularizer=regularizers.l1_l2(l1=wre_l1,l2=wre_l2),activation=hlr_act,input_shape=(x_train.shape[1],)))
            my_model.add(layers.Dropout(drp_out))
            my_model.add(layers.Dense(num_neu_hid,kernel_regularizer=regularizers.l1_l2(l1=wre_l1,l2=wre_l2),activation=hlr_act))
            my_model.add(layers.Dropout(drp_out))
            ##### Adding an extra hidden layer and the associated drop out
            extra_layer1=extra_layer1
            if extra_layer1==1:
                my_model.add(layers.Dense(num_neu_hid,kernel_regularizer=regularizers.l1_l2(l1=wre_l1,l2=wre_l2),activation=hlr_act))
                my_model.add(layers.Dropout(drp_out))
            ##### Here you can add an extra hidden layer with the associated drop out if you need it..
            my_model.add(layers.Dense(1,activation='sigmoid'))

            my_model.compile(optimizer=optimizers.RMSprop(lr=lr),loss=los_fun,metrics=[kpi])

            #print('Below the summary of the parameters processed :'+'\n'+str(my_model.summary()))
            from datetime import datetime
            k_start_time = datetime.now()
           
            globals()[i]= my_model.fit(xtr_dta,ytr_dta,epochs=num_epochs,batch_size=16, validation_data=(xvl_dta,yvl_dta),verbose=verbose)
            out=(globals()[i])
            
            # Storing the output of the model for each of the k folds:
            if i==0:
                out0=globals()[i]
            elif i==1:
                out1=globals()[i]
            elif i==2:
                out2=globals()[i]
            elif i==3:
                out3=globals()[i]
            #### Model fit time (for execution all folds execution):
                k_end_time = datetime.now()  
                global k_dur        
                k_dur=str(k_end_time - k_start_time)

        #### mean of validation accuracy for each fold
            globals()[i]=np.mean(out.history['val_acc'])

        #### standard deviation of the av. validation accuracy on each fold
        global val_acc_dispersion
        std_dev=np.array([globals()[0],globals()[1],globals()[2],globals()[3]])
        val_acc_dispersion=round(np.std(std_dev),4)

        #### overall mean of validation accuracy
        global av_validation_accuracy
        out0_val_acc=np.array(out0.history['val_acc'])
        out1_val_acc=np.array(out1.history['val_acc'])
        out2_val_acc=np.array(out2.history['val_acc'])
        out3_val_acc=np.array(out3.history['val_acc'])
        av_validation_accuracy=(out0_val_acc+out1_val_acc+out2_val_acc+out3_val_acc)/k
        
        global k_max_vld_acc
        k_max_vld_acc=np.argmax(av_validation_accuracy)+1
        
        global av_av_validation_accuracy
        av_av_validation_accuracy=round(np.mean(av_validation_accuracy),2)
        
        # average validation accuracy by fold:
        global f0_av_val_acc
        f0_av_val_acc=round(np.mean(out0.history['val_acc']),5)
        global f1_av_val_acc
        f1_av_val_acc=round(np.mean(out1.history['val_acc']),5)
        global f2_av_val_acc
        f2_av_val_acc=round(np.mean(out2.history['val_acc']),5)
        global f3_av_val_acc
        f3_av_val_acc=round(np.mean(out3.history['val_acc']),5)
                
        #### overall mean of training accuracy
        global av_training_accuracy
        out0_acc=np.array(out0.history['acc'])
        out1_acc=np.array(out1.history['acc'])
        out2_acc=np.array(out2.history['acc'])
        out3_acc=np.array(out3.history['acc'])
        av_training_accuracy=(out0_acc+out1_acc+out2_acc+out3_acc)/k
    
        global k_max_trn_acc
        k_max_trn_acc=np.argmax(av_training_accuracy)+1
        
        global av_av_training_accuracy
        av_av_training_accuracy=round(np.mean(av_training_accuracy),2)
        
        # average by fold:
        global f0_av_trn_acc
        f0_av_trn_acc=round(np.mean(out0.history['acc']),5)
        global f1_av_trn_acc
        f1_av_trn_acc=round(np.mean(out1.history['acc']),5)
        global f2_av_trn_acc
        f2_av_trn_acc=round(np.mean(out2.history['acc']),5)
        global f3_av_trn_acc
        f3_av_trn_acc=round(np.mean(out3.history['acc']),5)
        
        #### overall mean of validation loss
        global av_validation_loss
        out0_val_los=np.array(out0.history['val_loss'])
        out1_val_los=np.array(out1.history['val_loss'])
        out2_val_los=np.array(out2.history['val_loss'])
        out3_val_los=np.array(out3.history['val_loss'])
        av_validation_loss=(out0_val_los+out1_val_los+out2_val_los+out3_val_los)/k
    
        global k_min_vld_lss
        k_min_vld_lss=np.argmin(av_validation_loss)+1    
        
        global av_av_validation_loss
        av_av_validation_loss=round(np.mean(av_validation_loss),2)

        # average by fold:
        global f0_av_val_lss
        f0_av_val_lss=round(np.mean(out0.history['val_loss']),5)
        global f1_av_val_lss
        f1_av_val_lss=round(np.mean(out1.history['val_loss']),5)
        global f2_av_val_lss
        f2_av_val_lss=round(np.mean(out2.history['val_loss']),5)
        global f3_av_val_lss
        f3_av_val_lss=round(np.mean(out3.history['val_loss']),5)
        
        #### overall mean of training loss
        global av_training_loss
        out0_tra_los=np.array(out0.history['loss'])
        out1_tra_los=np.array(out1.history['loss'])
        out2_tra_los=np.array(out2.history['loss'])
        out3_tra_los=np.array(out3.history['loss'])
        av_training_loss=(out0_tra_los+out1_tra_los+out2_tra_los+out3_tra_los)/k

        global k_min_trn_lss
        k_min_trn_lss=np.argmin(av_training_loss)+1    
        
        global av_av_training_loss
        av_av_training_loss=round(np.mean(av_training_loss),2)
        
        # average by fold:
        global f0_av_trn_lss
        f0_av_trn_lss=round(np.mean(out0.history['loss']),5)
        global f1_av_trn_lss
        f1_av_trn_lss=round(np.mean(out1.history['loss']),5)
        global f2_av_trn_lss
        f2_av_trn_lss=round(np.mean(out2.history['loss']),5)
        global f3_av_trn_lss
        f3_av_trn_lss=round(np.mean(out3.history['loss']),5)
        
        #### Coefficient of correlation (linear, Pearson)
        
        epochs=range(1,len(av_training_loss)+1)
        
        global k_kco_val_acc
        k_kco_val_acc=round(np.corrcoef(epochs,av_validation_accuracy)[0,1],2)
        global k_kco_trn_acc
        k_kco_trn_acc=round(np.corrcoef(epochs,av_training_accuracy)[0,1],2)
        global k_kco_val_lss
        k_kco_val_lss=round(np.corrcoef(epochs,av_validation_loss)[0,1],2)
        global k_kco_trn_lss
        k_kco_trn_lss=round(np.corrcoef(epochs,av_training_loss)[0,1],2)
```

Another possible explanation is that smaller datasets have smaller intrinsic variance. So this means that your model properly captures patterns inside of your data and train error is greater simply because the inner variance of training set is greater then validation set.

## 4.2 CONCLUSIONS

Predicting customer behaviour is a frequently practiced exercise across businesess operating in numerous industries, not least the telecom industry, where churn predictions play a key role.

We have been trying to build an artifical neural network able to effectively address this problem, classifiable as binary classification.

The data used to perform the analysis have undergone heavy manipulation before being ready to be processed. 

One important aspect is the normalisation of the features, which can lead to better results. This becames especially true when the magnitudes of the predictors are far apart from each other.

Selecting the right learning rate has proven essential to converge towards a high accuracy while building a reliable model. In the specific case, the shift from the initial 1% learning rate to 0.1% has had a remarkable impact over the model performance.

Doubts remain about the the validation set outperforming the training set quite regularly - despite the difference being small. A possible explanation is that smaller datasets have smaller intrinsic variance. 
The k-fold cross-validation has been performed to investigate the potential unrepresentativeness of the validation set. This hasn't outlined any critical issue related to the validation data extraction.
The data, downloaded from the IBM website, might have been freely shared to allow a fairly easy modelling for demonstration purposes. It it has in fact been fairly easy to achieve a high accuracy in the validation set, once the data normalisation was completed.

With such a high accuracy and very little evidence of data over fitting, it hasn't seemed to be the case to experiment techniques used to tackle over fitting, such as weights regularisation and drop-out. 

However, reducing the size of the network has in some cases proven to return better results. This specifically refers to the worse results generated by the ANNs with +1 hidden layer (see model 11, 12). Probably a network with 1 hidden layer is the best architecture choice to address this binary classification problem.

The model run on the test data set returns an average of ~96% of correctly classified events -  extremely close to the vlaues 
observed during the validation stage. With such a good result, we can have good confidence that the netowrk will do a good job in helping tackle customers churn.
